## PREPARE THE DATA BEFORE APPLYING FACEBOOK PROPHET TOOL 

```python
avocado_df
```

```python
avocado_profit_df = avocado_df[['Date', 'AveragePrice']]
avocado_profit_df
```

```python
avocado_prophet_df = avocado_profit_df.rename(columns= {'Date': 'ds', 'AveragePrice':'y'})
```

```python
avocado_prophet_df
```

## UNDERSTAND INTUITION BEHIND FACEBOOK PROPHET

## DEVELOP MODEL AND MAKE PREDICTIONS - PART A

```python
m = Prophet()
m.fit(avocado_prophet_df)
```

```python
future = m.make_future_dataframe(periods = 365)
forecast = m.predict(future)
```

```python
forecast.head(10)
```

```python
figure = m.plot(forecast, xlabel = 'Date', ylabel = 'Price')
```

```python
figure2 = m.plot_components(forecast)
```

## DEVELOP MODEL AND MAKE PREDICTIONS (REGION SPECIFIC) - PART B

```python
# dataframes creation for both training and testing datasets 
avocado_df = pd.read_csv('/content/gdrive/My Drive/Predicting Future Product Prices/avocado.csv')
```

```python
# Select specific region
avocado_df_sample = avocado_df[avocado_df['region']=='West']
```

```python
avocado_df_sample = avocado_df_sample.sort_values('Date')
```

```python
avocado_df_sample
```

```python
plt.plot(avocado_df_sample['Date'], avocado_df_sample['AveragePrice'])
```

```python
avocado_df_sample = avocado_df_sample.rename(columns = {'Date':'ds', 'AveragePrice':'y'})
```

```python
avocado_df_sample
```

```python
m = Prophet()
m.fit(avocado_df_sample)
# Forcasting into the future
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
```

```python
figure = m.plot(forecast, xlabel='Date', ylabel='Price')
```

```python
figure3 = m.plot_components(forecast)
```
